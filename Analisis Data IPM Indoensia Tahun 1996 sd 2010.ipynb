# 1. Deskripsi Data

Himpunan data yang akan dianalis di sini adalah data Indeks Pembangunan Manusia (IPM) per Provinsi di seluruh Indonesia. IPM adalah pengukuran perbandingan dari harapan hidup, melek huruf, pendidikan dan standar hidup. Anga IPM digunakan untuk mengklasifikasikan apakah sebuah negara adalah negara maju, negara berkembang, atau negara terbelakang. Selain itu, nilai IPM dapat juga digunakan untuk mengukur pengaruh dari kebijaksanaan ekonomi terhadap kualitas hidup.


Dataset yang digunakan di sini adalah data IPM seluruh Provinsi di seluruh Indonesia. Penjelasan mengenai Variabel pada Dataset yang digunakan di sini adalah sebagai berikut.

1. kode_provinsi: Kode Provinsi
2. nama_provinsi: Nama Provinsi
3. tahun: Tahun
4. indeks_pembangunan_manusia: Indeks Pembangunan Manusia

Perserikatan Bangsa-Bangsa (PBB) menetapkan peringkat kinerja pembangunan manusia pada skala 0,0 – 100,0 berdasarkan nilai IPM. Berikut ini adalah kategori sebagai berikut yang merepresentasikan nilai IPM dan kategori indeks pembangunan manusia:

1. Tinggi : IPM > 80,0
2. Menengah Atas : IPM antara 66,0 – 79,9
3. Menengah Bawah : IPM antara 50,0 – 65,9
4. Rendah : IPM < 50,0

Data IPM ini diperoleh dari https://data.go.id/dataset/indeks-pembangunan-manusia-ipm-per-provinsi


# 2. Statistik Tentang Data

```python
#mengambil data dari file csv
import pandas as pd
data_ipm = pd.read_csv('ipm-per-provinsi-1996-2010.csv', delimiter = ',', encoding = "ISO-8859-1")
```

```python
#menampilkan data
data_ipm
```

```python
#menampilkan banyaknya data yang terdapat pada file csv
data_ipm.count()
```

```python
#menampilkan data yang dikelompokkan per provinsi
data_ipm.groupby('nama_provinsi').mean()
```

```python
#menampilkan data yang dikelompokkan per provinsi
data_ipm.groupby('tahun').mean()
```

# 3. Visualisasi Sederhana

```python
#menampilkan rata-rata indeks pembangunan manusia berdasarkan provinsi dari tahun 1996 s.d 2010
ax = data_ipm.groupby('nama_provinsi').mean()['indeks_pembangunan_manusia']
ax = ax.plot(kind='bar', title ="Rata-Rata IPM Berdasarkan Provinsi", legend=False, fontsize=12, figsize=(15,10))
ax.set_xlabel("Provinsi", fontsize=12)
ax.set_ylabel("Indeks Pembangunan Manusia", fontsize=12)
```

```python
#impor modul dari library matplotlib
from matplotlib import pyplot as plt

#menampilkan indeks pembangunan manusia berdasarkan tahun
ax = data_ipm.groupby('tahun').mean()['indeks_pembangunan_manusia']
ax = plt.plot(ax)
plt.xlabel("Tahun", fontsize=12)
plt.ylabel("Indeks Pembangunan Manusia", fontsize=12)
```
