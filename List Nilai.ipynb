```python
import random 
#generate random
nilaiTugas =[]
for i in range(250):
    nilaiTugas.append(random.randrange(1,100,1))
    
print(nilaiTugas)
```

```python
#menampilkan bilangan ganjil
print("Data semua nilai tugas yang merupakan bilangan ganjil:")
for n in nilaiTugas:
    if(n%2==1):
        print (n)
```

```python
#Mencari banyaknya nilai yg lebih dari 70
hitung = 0
for x in nilaiTugas:
    if x>=70:
        hitung +=1
        print (x)
        print ("Nilai >=70 ke-", hitung)
```

```python
#Menghitung Standar Deviasi

#mencari rata-rata
total = 0
for n in nilaiTugas:
    total= total + n
rata2 = total/len(nilaiTugas)

#standar deviasi (sd)
selisih_total = 0
for n in nilaiTugas:
    x = n-rata2
    pangkat=x**2
    selisih_total=selisih_total + pangkat
sd=(selisih_total/(len(nilaiTugas)-1))**(0.5)
print("Standard Deviasi=", sd)
```

```python
#Nilai Maksimum dan Minimum
maks=0
for n in nilaiTugas:
    if n>maks:
        maks=n
print("Nilai Maksimum=",maks)

mins=99
for n in nilaiTugas:
    if n<mins:
        mins=n
print("Nilai Minimum=", mins)
```
